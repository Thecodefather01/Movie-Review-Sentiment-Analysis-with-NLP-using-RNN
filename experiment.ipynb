```python
import tensorflow as tf
import keras
```

```python
import pandas as pd
import sklearn
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ["the glass of milk",
        "the glass of juice",
        "the cup of tea",
        "I am a good boy",
        "I am a good developer",
        "understand the meaning of words",
        "your videos are good"]
```

```python
voc_size=10000 #vector range
```

```python
one_hot_repr = [one_hot(words,voc_size)for words in sent]
one_hot_repr
```

```python
## word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
sent_len = 8
#making every sentence equal to max length - 8 by adding 0's to pre or post sentence vectors
embedded_docs = pad_sequences(one_hot_repr,padding="pre",maxlen=sent_len)
embedded_docs
```

```python
##feature representation
dim=10
```

```python
model=Sequential()

model.add(Embedding(input_dim=voc_size, output_dim=dim))
model.build(input_shape=(None, sent_len))
model.compile(optimizer="adam", loss="mse")
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
